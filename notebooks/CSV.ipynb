```python
import os
import pandas as pd
import re

p = "../output/SeparationSimulator/10-19-profile/3-spread-smallf/Strict/"
dir_list = next(os.walk(p))[1]

dataframes = []
for dirr in dir_list:
    if "lambda-1.00" in dirr:
        a = re.search('alpha-(.*)', dirr)
        alpha = a.group(0)
        csv = os.path.join(p, dirr, "metrics.csv")
        df = pd.read_csv(csv, ";", index_col=0)
        df.rename(columns = {'Homogeneous neighborhoods':alpha}, inplace = True)
        dataframes.append(df[alpha])
```

```python
df = pd.concat(dataframes, axis=1)
df.to_csv("combined-smallf.csv", ";")
```
